**9. Processed Raw data file**

The raw data file is alabama16_6rigions_ms.sas7bdat.cvs, Alabama14_ms.xlsx, Alabama12_ms.xlsx, which are on the same drive.

Combining Columns CR4, CR5A to CR5E to merge it to CR5:
If a row has a value in "CR5A", set the value in "CR5" to 1. If a row has a value in "CR5B", set the value in "CR5'' to 2. Because CR5A is Indian based on the file Alabama YTS 2016-MS codebook,it corresponds to the value 1, which means Indian on our race map.

Similarly, "CR5C" corresponds to 3, "CR5D" to 4, and "CR5E" to 5.

CR4 is a survey question ask the person whether or not is Hispanic. 1 is no. 2,3,4,5 each means specifc Hispanic group. I will set the value in my "CR5" to 6, which means Hispanic on my map.

If a row has values in more than one of the columns "CR5A" to "CR5E", set the value in "CR5" to 7, which means two races or more on our map.

Selecting Specific Columns:
From the filtered rows, select the following columns for gender, grade and race:  "CR2", "CR3", "CR5".

```python
from ctgan import CTGAN
def merge_cr5(row):
    cr5_cols = ['CR5A', 'CR5B', 'CR5C', 'CR5D', 'CR5E']
    cr5_flags = [1, 2, 3, 4, 5]
    cr5_values = [not pd.isna(row[col]) for col in cr5_cols]  # generates a Boolean list indicating whether each column has a value

    if sum(cr5_values) > 1:  # 7 represents two or more races
        return 7
    elif row['CR4'] in [2, 3, 4, 5] and any(cr5_values):
        return 7
    elif sum(cr5_values) == 1:
        return cr5_flags[cr5_values.index(True)]
    elif row['CR4'] in [2, 3, 4, 5]:
        return 6
    else:
        return None

# Preprocess 2012 data
data_2012 = pd.read_excel('/content/drive/My Drive/Alabama12_ms.xlsx')
data_2012 = data_2012.copy()
data_2012['CR5'] = data_2012.apply(merge_cr5, axis=1)
selected_columns = ['CR2', 'CR3', 'CR5', 'CR7', 'CR11', 'CR12', 'CR13', 'CR21', 'CR22', 'CR23', 'CR26', 'CR27', 'CR28', 'CR54', 'CR56']
data_2012 = data_2012.loc[:, selected_columns]

data_2014 = pd.read_excel('/content/drive/My Drive/Alabama14_ms.xlsx')
data_2014 = data_2014.copy()
data_2014['CR5'] = data_2014.apply(merge_cr5, axis=1)
data_2014 = data_2014.loc[:, selected_columns]

data_2016 = pd.read_csv('/content/drive/My Drive/alabama16_6rigions_ms.sas7bdat.csv')
data_2016 = data_2016.copy()
data_2016['CR5'] = data_2016.apply(merge_cr5, axis=1)
data_2016 = data_2016.loc[:, selected_columns]

# Load the 2024 input data
inputs_2024 = pd.read_csv('/content/drive/My Drive/2024_sample_data_not_survey.csv')
required_columns = ['CR2', 'CR3', 'CR5']
# Ensure the 2024 input data columns match the required columns
conditions_2024 = inputs_2024.loc[:, required_columns]
data_2016
```

**10. generate synthetic survey answers for 2024 Montgomery data by CTGAN**

I choose to use CTGAN model to help generate synthetic data. CTGAN is a collection of Deep Learning based synthetic data generators for single table data, which are able to learn the relationship from real data and generate synthetic data with high fidelity.

I combine the processed data data_2012, data_2014, data_2016 to train our CTGAN model. Then I use the required column CR2(gender), CR3(grade) and CR5(race) in 2024_sample_data_not_survey.csv.csv as our input condition. Then I can generate the synthetic survey answers for "CR7", "CR11", "CR12", "CR13", "CR21", "CR22", "CR23", "CR26", "CR27", "CR28", "CR54", and "CR56".

If there are some missing values of CR2,CR3 and CR5 in 2024_sample_data_not_survey.csv, I just append the original row of the data_2016 in our sample.

At last, I combine the synthetic responses with the original input data and store the sample data in the file called synthetic_Montgomery_data_2024.csv.  I store it in the same google drive.

But CTGAN model has its limitations. I can not show the time effect towards the synthesis dataset, like the difference among dataset in different years. The only input is the students' race, gender and grade distribution in 2024.


```python
def CTGAN_model(data_2012, data_2014, data_2016, selected_columns, conditions_2024, required_columns):
    # Combine all preprocessed data
    combined_data = pd.concat([data_2012, data_2014, data_2016], ignore_index=True)
    # Initialize and train the CTGAN model
    ctgan = CTGAN()
    ctgan.fit(combined_data, discrete_columns=selected_columns)
    # Generate a large pool of synthetic data
    synthetic_pool = ctgan.sample(len(inputs_2024) * 10)

    # Function to filter the synthetic pool based on conditions
    def filter_synthetic_data(pool, conditions):
        for column, value in conditions.items():
            pool = pool[pool[column] == value]
        if pool.empty:
            raise ValueError("No matching data found for the given conditions.")
        return pool.sample(n=1, replace=True).drop(columns=list(conditions.keys()))

    # Generate synthetic responses for 2024 inputs
    synthetic_responses = []

    for _, row in conditions_2024.iterrows():
        conditions = row.to_dict()
        try:
            synthetic_sample = filter_synthetic_data(synthetic_pool, conditions)
            synthetic_responses.append(synthetic_sample)
        except ValueError:
            # For cases where no matching data can be found, append the original row
            print(f"No matching data found for conditions: {conditions}")
            synthetic_responses.append(pd.DataFrame([row]))

    # Convert list of synthetic responses to DataFrame
    synthetic_responses_df = pd.concat(synthetic_responses, ignore_index=True)
    # drop the race, grade, and gender columns from the synthetic responses before merging
    synthetic_responses_df = synthetic_responses_df.drop(columns=required_columns, errors='ignore')
    # Combine synthesized response with the original 2024 input data, retaining additional attributes
    synthetic_data_2024 = pd.concat([inputs_2024.reset_index(drop=True), synthetic_responses_df.reset_index(drop=True)], axis=1)

    # Save the synthetic data
    synthetic_data_2024.to_csv('synthetic_data_2024.csv', index=False)

CTGAN_model(data_2012, data_2014, data_2016, selected_columns, conditions_2024, required_columns)
```
